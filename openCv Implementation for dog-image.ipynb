```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def ImageShow(imageTitle, image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(imageTitle)
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```

# Home Work On Both saving and showing 25 images with (5*5) Dimensions

```python
imgeData3= cv2.imread("../Classes/dataset/training_set/dogs/dog.1001.jpg")
```

```python
ImageShow('dog-image',imgeData3)
```

```python
imgeData3.shape
```

```python
size =(weight,height)
angle = 2
imageNumber=0
for i in range(25):
    if i < 10:
        mgr = cv2.getRotationMatrix2D(size,angle+(i*1.5),1)
        rotation = cv2.warpAffine(imgeData3, mgr, size)
        cv2.imwrite("../ImagesSaved/"+"dog-"+str(imageNumber)+".jpg",rotation)
        imageNumber = imageNumber + 1
        
    else:
        mgr = cv2.getRotationMatrix2D(size,angle - (i*1.5),1)
        rotation = cv2.warpAffine(imgeData3, mgr, size)
        imageNumber = imageNumber + 1
        cv2.imwrite("../ImagesSaved/"+"dog-"+str(imageNumber)+".jpg",rotation)
        imageNumber = imageNumber + 1
    
```

```python
def LoadImageFromDirectory():
    
    path = ("../ImagesSaved/")
    
    fig = plt.figure(figsize = (10,10))
    i=1
    for file in os.listdir(path):
        pic = cv2.imread(os.path.join(path,file))
        pic = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
        pic = cv2.resize(pic,(224,224))
        
        plt.subplot(5,5,i)
        plt.xticks([])
        plt.yticks([])
        plt.grid([True])
       
        plt.tight_layout()
        plt.title(file)
        plt.imshow(pic)
        i=i+1
    plt.show()
```

```python
LoadImageFromDirectory()
```
